# Sandardization of X variable

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_std=df[['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']]
x_std=sc.fit_transform(x_std)
x_std
```

```python
x_std =pd.DataFrame(x_std,columns=['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year'])
x_std
```

```python
x.isnull().sum()
```

```python
x[['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']]=x_std
x
```

```python
x.dropna(inplace=True)
x
```

# Test Train Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Training the Model

```python
import sklearn
```

```python
pip install scikit-learn
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor(random_state=2529)
rfr.fit(x_train,y_train)
```

# Model Prediction

```python
y_pred=rfr.predict(x_test)
y_pred.shape
```

```python
y_pred
```

# Model Evaluation

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```
